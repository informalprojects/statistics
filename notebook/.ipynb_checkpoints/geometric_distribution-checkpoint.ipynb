```python
# Assuming each trial is independent
# Given probability of success
# To ensure 95% chance to have at least 1 success in n trials
# n = ln(0.05) / ln(1 - p)
```

```python
# Geometric distribution
# Probability of the first success happening at the k-th trial
# y = (1 - p)^(k - 1) * p
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def geometric(p, k):
    geo = ((1 - p)**(k - 1)) * p
    return geo
```

```python
x = np.arange(1,10, 1)
y = [geometric(0.5, x_i) for x_i in x]
```

```python
plt.plot(x, y)
```

```python
def geometric_cdf(p, k):
    geo_cdf = 1 - (1 - p) ** k
    return geo_cdf
```

```python
y_cdf = [geometric_cdf(0.5, x_i) for x_i in x]
```

```python
plt.plot(x, y_cdf)
```

```python
def success_threshold(p, t):
    trials = np.log(1 - t) / np.log(1 - p)
    return trials
```

```python
success_threshold(0.03, 0.99)
```

```python
p_range = np.arange(0.1, 1, 0.1)
t_range = np.arange(0.1, 1, 0.1)
```

```python
sp_range = [success_threshold(p_i, 0.9) for p_i in p_range]
st_range = [success_threshold(0.1, t_i) for t_i in t_range]
```

```python
plt.plot(p_range, sp_range)
```

```python
plt.plot(t_range, st_range)
```
